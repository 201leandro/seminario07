### 02. Criando um banco de dados PostgreSQL

#### A. Nos seus programas, procure e abra o pgAdmin 4, que foi instalado junto com o PostgreSQL

#### B. No menu esquerdo, abra o grupo "Servers", depois o servidor local "PostgreSQL 10", em seguida "Databases"

#### C. Clique com o botão direito do mouse sobre o menu "Databases", então "Create", e por fim "Database"

#### D. Dê um nome para o seu banco de dados e clique em "Save" para criá-lo. Aqui, daremos o nome de "mydb" (My DataBase).

### 03. Criando uma tabela

#### A. Expanda o menu do banco de dados: "mydb" > "Schemas" > "puclic" > "Tables"

#### B. Clique com o botão direito sobre o menu "Tables" > "Create" > "Table..."

#### C. Dê um nome pra tabela. Aqui, daremos o nome de "mytable". Após isso, clique em "Columns", no menu superior.

#### D. No menu "Columns" é onde criamos as colunas da nossa tabela, de acordo com os dados que iremos armazenar. Para adicionar uma coluna, clique no botão de "+" no canto superior direito.

#### E. Para criar uma coluna devemos preencher
* nome da coluna
* o tipo de dado que ela irá armazenar
* o tamanho (caso o tipo de dado necessite)
* a precisão (caso o tipo de dado necessite)
* se a coluna aceitará valores vazios
* se a coluna é uma chave primária

* **Boolean**: Armazena "true" ou "false" (ou 1 para verdadeiro e 0 para falso).
* **Character varying**: Armazena texto (necessita informar o Length, que será o tamanho máximo permitido).

#### F. Clique em "Save" para criar a tabela.

### 04. Importando dados de um arquivo .csv para a tabela

#### A. Clique com o botão direito sobre a tabela, em seguida em "Import/Export".

#### B. Abrirá o assitente de importação/exportação de dados para tabelas. Devemos informar:

* Import ou Export
 * Se desejamos importar ou exportar dados da tabela selecionada. Marque "import".
* Filename
 * O caminho do arquivo de dados que será importado.
* Format
 * A extensão do arquivo que será importado. Escolha "csv".
* Enconding
 * A codificação usada no processo. Escolha "UTF-8"
* Header
 * Se o arquivo que será importado possui cabeçalho, ou seja, se a primeira linha do csv são os nomes das colunas.
* Delimiter
 * O caracter separador de colunas no csv. No nosso caso, é a própria vírgula ",".
* Quote
 * O tipo de aspas que caracteriza um texto no dataset. No nosso caso são as aspas duplas.

**As configurações não mencionadas podem ser deixadas no padrão**

#### C. Clique em "OK".

**Observações importantes:**
* As colunas no dataset devem estar de acordo com o tipo de dado e com a formatação definida na criação da coluna
* Caso seja utilizado cabeçalho no csv, este deve estar de acordo com os nomes das colunas na tabela
* A chave primária deverá estar presente no csv, ou excluida da listagem das colunas que serão importadas
* Caso haja algum erro de importação, o pgAdmin informa o erro no modal de importação, e cancela todo o processo
